```python
test_df = pd.read_csv('D:\\Prasad Personal\\kaggle_competitions\\titanic\\test.csv')
test_df
```

```python
def preprocessing(df):
    ports = pd.get_dummies(df.Embarked, prefix='embarked')
    df = df.join(ports)
    df['Sex'] = df.Sex.map({'male':1,'female':0})
    df.Age.fillna(df.Age.mean(), inplace=True)
    df.Fare.fillna(df.Fare.mean(), inplace=True)
    df.drop(['Name', 'PassengerId','Ticket', 'Cabin','Embarked'], axis=1, inplace=True)
    return df
```

```python
preprocessed_df = preprocessing(test_df)
```

```python
preprocessed_df.isnull().sum()
```

```python
predicted_results = model.predict(preprocessed_df)
```

```python
my_submission = pd.DataFrame({'PassengerId': test_df.PassengerId, 'Survived': predicted_results})
# you could use any filename. We choose submission here
my_submission.to_csv('D:\\Prasad Personal\\kaggle_competitions\\titanic\\submission_v1.csv', index=False)
```
